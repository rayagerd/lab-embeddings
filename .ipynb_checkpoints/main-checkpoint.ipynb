## Task 1: Semantic Similarity Comparison
### Objective:
Compare semantic similarity between pairs of sentences using cosine similarity and embeddings.

### Steps:
1. Load a pre-trained Sentence Transformer model.
2. Encode the sentence pairs.
3. Compute cosine similarity for each pair.

### Dataset:
- "A dog is playing in the park." vs. "A dog is running in a field."
- "I love pizza." vs. "I enjoy ice cream."
- "What is AI?" vs. "How does a computer learn?"


```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sentence pairs
sentence_pairs = [
    ("A dog is playing in the park.", "A dog is running in a field."),
    ("I love pizza.", "I enjoy ice cream."),
    ("What is AI?", "How does a computer learn?")
]

# Compute similarities
for sentence1, sentence2 in sentence_pairs:
    embeddings = model.encode([sentence1, sentence2])

    similarity = cosine_similarity([embeddings[0]], [embeddings[1]])
    similarity_score = similarity[0][0]

    print(f"Sentence 1: {sentence1}")
    print(f"Sentence 2: {sentence2}")
    print(f"Cosine Similarity: {similarity_score:.4f}")
    print("-" * 50)

#YOUR CODE HERE
```

### Questions:
- Which sentence pairs are the most semantically similar? Why?
The most semantically similar pair is “I love pizza.” and “I enjoy ice cream.” because it has the highest cosine similarity score. Both sentences express a positive preference toward food. The dog sentence pair is also very similar, with a score of 0.5220, because both sentences describe a dog doing an outdoor activity. Since the difference between these two scores is very small, they can both be considered highly similar.

- Can you think of cases where cosine similarity might fail to capture true semantic meaning?
Cosine similarity can fail when sentences have similar words but opposite meanings, such as “I like this movie” and “I do not like this movie.” It can also struggle with sarcasm, missing context and ambiguous words. Sincee cosine similarity compares vector direction, it does not fully understand meaning the way humans do.

## Task 2: Document Clustering
### Objective:
Cluster a set of text documents into similar groups based on their embeddings.

### Steps:
1. Encode the documents using Sentence Transformers.
2. Use KMeans clustering to group the documents.
3. Analyze the clusters for semantic meaning.

```python
from sklearn.cluster import KMeans

# Documents to cluster
documents = [
    "What is the capital of France?",
    "How do I bake a chocolate cake?",
    "What is the distance between Earth and Mars?",
    "How do I change a flat tire on a car?",
    "What is the best way to learn Python?",
    "How do I fix a leaky faucet?"
]

model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode documents

#YOUR CODE HERE
doc_embeddings = model.encode(documents)
print(type(doc_embeddings))
print(doc_embeddings.shape)
print(doc_embeddings[:2])
```

```python
# Perform KMeans clustering

#YOUR CODE HERE
kmeans = KMeans(n_clusters=3,random_state=100, n_init=10)
kmeans.fit(doc_embeddings)
pred = kmeans.predict(doc_embeddings) 
print(pred)
```

```python
# Print cluster assignments

#YOUR CODE HERE
for cluster_id in range(3):
    print(f"\nCluster {cluster_id}:")
    
    for document, cluster in zip(documents, pred):
        if cluster == cluster_id:
            print("-", document)
```

### Questions:
- How many clusters make the most sense? Why?
- Examine the documents in each cluster. Are they semantically meaningful? Can you assign a semantic "theme" to each cluster?
- Try this exercise with a larger dataset of your choice

The documents were grouped into three clusters based on their sentence embeddings. 

Cluster 0 contains practical questions and has a clear semantic meaning because both documents are about fixing physical problems.

Cluster 1 contains a mix of factual and instructional questions. This cluster is less semantically clear, which shows that KMeans can produce imperfect groupings, especially with a very small dataset.

Cluster 2 contains only the Python learning question. This suggests that the model considered it semantically different from the other documents.

Overall, the embeddings allowed KMeans to group some similar documents together, but the clustering is limited because there are only six documents.

## Task 3: Semantic Search System
### Objective:
Create a semantic search engine:
A user provides a query and you search the dataset for semantically relevant documents to return. Return the top 5 results.

### Dataset:
- Use the following set of documents:
    - "What is the capital of France?"
    - "How do I bake a chocolate cake?"
    - "What is the distance between Earth and Mars?"
    - "How do I change a flat tire on a car?"
    - "What is the best way to learn Python?"
    - "How do I fix a leaky faucet?"
    - "What are the best travel destinations in Europe?"
    - "How do I set up a local server?"
    - "What is quantum computing?"
    - "How do I build a mobile app?"


```python
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

# Documents dataset
documents = [
    "What is the capital of France?",
    "How do I bake a chocolate cake?",
    "What is the distance between Earth and Mars?",
    "How do I change a flat tire on a car?",
    "What is the best way to learn Python?",
    "How do I fix a leaky faucet?",
    "What are the best travel destinations in Europe?",
    "How do I set up a local server?",
    "What is quantum computing?",
    "How do I build a mobile app?"
]

# Compute document embeddings

#YOUR CODE HERE
doc_embeddings = model.encode(documents)
print(type(doc_embeddings))
print(doc_embeddings.shape)
print(doc_embeddings[:2])
```

```python
# Create the search function
#This function should encode the user query and return the top N documents that most resemble it
def semantic_search(query, documents, doc_embeddings, top_n=5):
    # YOUR CODE HERE
    query_embedding = model.encode([query])
    similarities = cosine_similarity(query_embedding, doc_embeddings)
    scores = similarities[0]
    top_scores = scores.argsort()[::-1][:top_n]
    results = []
    
    for score in top_scores:
        results.append((documents[score], scores[score]))
    return results
```

```python
# Test the search function
query = "Explain programming languages."
results = semantic_search(query, documents, doc_embeddings)

for document, score in results:
    print(f"Score: {score:.4f} | Document: {document}")
```
